# EDA

```python
class_counts = df['Class'].value_counts()

labels = ['Non-Fraud', 'Fraud']
colors = ['Green', 'Red']

plt.figure(figsize=(4, 4))
plt.pie(
    class_counts, 
    labels=[f'{label} ({count})' for label, count in zip(labels, class_counts)],  # Include counts in labels
    startangle=90, 
    colors=colors, 
    explode=[0, 0.1],
    autopct='%1.1f%%'  # Show percentages
)
plt.title("Class Distribution: Fraud vs. Non-Fraud")
plt.show()
```

```python
plt.figure(figsize=(10, 6))

# Separate the classes for better visibility
class_0 = df[df['Class'] == 0]
class_1 = df[df['Class'] == 1]

# Plot class 0
plt.scatter(class_0['Time'], class_0['Amount'], c='green', alpha=0.5, label='Class 0', s=50)

# Plot class 1 with larger markers and distinct color
plt.scatter(class_1['Time'], class_1['Amount'], c='red', alpha=1.0, label='Class 1', s=50, edgecolors='black')

plt.title('Transaction Amounts Over Time')
plt.xlabel('Time (seconds since first transaction)')
plt.ylabel('Amount')
plt.legend()
plt.show()
```

All frauds are of amount between 0 and 3000

```python
# Correlation matrix
correlation_matrix = df.corr()

# Visualize the correlation matrix
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, cmap='coolwarm', annot=False, fmt=".2f")
plt.title('Feature Correlations')
plt.show()
```

```python
scaler = StandardScaler()

# Scale 'Amount' and 'Time' features
df['Time'] = scaler.fit_transform(df['Time'].values.reshape(-1, 1))
df['Amount'] = scaler.fit_transform(df['Amount'].values.reshape(-1, 1))
```

Balance Datasets

```python
from imblearn.over_sampling import SMOTE

# Split features and target
X = df.drop('Class', axis=1)
y = df['Class']

# Apply SMOTE
smote = SMOTE(random_state=42)
X_resampled, y_resampled = smote.fit_resample(X, y)

# Visualize the balanced class distribution
sns.barplot(x=y_resampled.value_counts().index, y=y_resampled.value_counts().values)
plt.title('Balanced Class Distribution')
plt.xlabel('Class')
plt.ylabel('Count')
plt.show()
```

Training and Test data Split

```python
# Train-test split
X_train, X_test, y_train, y_test = train_test_split(
    X_resampled, y_resampled, test_size=0.2, random_state=42, stratify=y_resampled
)

# Print shapes to confirm split
print("Training data shape:", X_train.shape)
print("Testing data shape:", X_test.shape)
```

# Baseline Models


**Logistic Regression**

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix, classification_report, roc_auc_score, roc_curve

# Initialize the model
logistic_model = LogisticRegression(max_iter=1000, random_state=42)

# Train the model
logistic_model.fit(X_train, y_train)
```

```python
# Predictions
y_pred = logistic_model.predict(X_test)
y_proba = logistic_model.predict_proba(X_test)[:, 1]

# Confusion Matrix
print("Confusion Matrix (Logistic Regression):")
print(confusion_matrix(y_test, y_pred))

# Classification Report
print("Classification Report (Logistic Regression):")
print(classification_report(y_test, y_pred))

# AUC-ROC Score
roc_auc = roc_auc_score(y_test, y_proba)
print(f"AUC-ROC Score (Logistic Regression): {roc_auc:.4f}")

# AUC-ROC Curve
fpr, tpr, thresholds = roc_curve(y_test, y_proba)
plt.figure(figsize=(8, 6))
plt.plot(fpr, tpr, label=f"Logistic Regression (AUC = {roc_auc:.4f})")
plt.plot([0, 1], [0, 1], 'r--')  # Diagonal line
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('AUC-ROC Curve')
plt.legend()
plt.show()
```

Feature Importance

```python
from sklearn.ensemble import RandomForestClassifier

# Fit a Random Forest model to get feature importance
rf_model = RandomForestClassifier(random_state=42)
rf_model.fit(X_train, y_train)

# Get feature importance
feature_importances = pd.DataFrame({
    'Feature': X_train.columns,
    'Importance': rf_model.feature_importances_
}).sort_values(by='Importance', ascending=False)

# Plot feature importance
plt.figure(figsize=(10, 6))
sns.barplot(x='Importance', y='Feature', data=feature_importances.head(10))
plt.title('Top 10 Features Indicating Fraud')
plt.show()
```

```python
import shap
from sklearn.linear_model import LogisticRegression

# 1. Assume X_train, X_test, y_train, y_test, and 'df.columns' are ready.
feature_names = X_train.columns

# 2. Train logistic regression
logistic_model = LogisticRegression(max_iter=1000, random_state=42)
logistic_model.fit(X_train, y_train)

# 3. Create SHAP explainer
explainer = shap.Explainer(logistic_model, X_train, feature_names=feature_names)

# 4. Get SHAP values for test data
shap_values = explainer(X_test)

# 5. Summary Plot (global importance)
shap.plots.beeswarm(shap_values)

# 6. Bar plot (mean absolute shap)
shap.plots.bar(shap_values)

# 7. Dependence Plot (zoom in on a single feature, e.g. V14)
shap.plots.scatter(shap_values[:, "V14"], color=shap_values)

# 8. Explain an individual prediction (sample index 0)
shap.initjs()
shap.plots.force(shap_values[0])
```
